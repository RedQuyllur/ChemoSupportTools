## **Part I** Data read & format conversion

At first we need to load the dataset and merge with suplementary data. After that data should be converted to project format.

```python
url = "https://raw.githubusercontent.com/QIBChemometrics/Benchtop-NMR-Coffee-Survey/main/Benchtop_NMR_Coffee_Survey.csv"
dataset_raw = pd.read_csv(url)

axis_label = "Chemical shift (ppm)"
axis_chem_shift = dataset_raw.loc[:,axis_label].values.tolist()
dataset_raw = dataset_raw.drop(axis_label, axis=1)

dataset_processed = pd.DataFrame(columns=["Sample Code", CONSTANTS.X_CHEM_SHIFT, CONSTANTS.Y_SPECTRUM])
for sample in dataset_raw.columns[1:]:
    spectrum = dataset_raw.loc[:,sample].values.tolist()
    sample_dict = ({
        "Sample Code": sample,
        CONSTANTS.X_CHEM_SHIFT: [axis_chem_shift],
        CONSTANTS.Y_SPECTRUM: [spectrum],
    })
    dataset_processed = pd.concat([dataset_processed, pd.DataFrame(sample_dict)])
dataset_processed.reset_index(drop=True, inplace = True)

suplementary_data_file_loc = path.join(Paths.DATASETS.get("raw"),"benchtop_nmr_coffee_survey_spectra","suplementary_table.csv")
suplementary_data = pd.read_csv(suplementary_data_file_loc)

dataset_processed = pd.merge(suplementary_data, dataset_processed)

dataset_to_save_pkl = {"fold0": dataset_processed}

config_local = {
    "root_dir_load": None,
    "root_dir_save": path.join(Paths.DATASETS.get("processed"), "nmr_spectra"),
    "file_name": "coffee_survey_spectra_prepared",
    "file_format": ".pkl",
}

dataset_description={
    "format_data": "raw",
    "format_version": None,
    "comment": "Raw dataset of coffee survey NMR spectra."
}

reader = BasicReader(
    load_data_on_init=False,
    load_data_on_init_source="local",
    configuration_local=config_local
)
reader.data = dataset_to_save_pkl
reader.data_descriptor = dataset_description
reader.data_save_to_file()
```

## **Part II** Basic preprocessing and data nomralization

```python
axis_chem_shift_range = [3.1, 3.7]
axis_chem_shift = dataset.get("axis_chem_shift")[0]

def closest(lst, value):
    return min(range(len(lst)), key = lambda i: abs(lst[i]-value))

axis_chem_shift_range = tuple([closest(axis_chem_shift, shift) for shift in axis_chem_shift_range])

data_format = {
    "unpack": {
        CONSTANTS.PROCESS_DATA_AS_MULTIPLE_YX: {
            "y": [CONSTANTS.Y_SPECTRUM, CONSTANTS.X_CHEM_SHIFT]
        }},
    "pack": {
        CONSTANTS.PROCESS_DATA_AS_MULTIPLE_YX: {
            "y": [CONSTANTS.Y_SPECTRUM, CONSTANTS.X_CHEM_SHIFT]
        }}
}

closeup_region = Preprocessor.dimensions.cut_off_the_range(
    data=dataset_pkl, class_constants=CONSTANTS, data_format=data_format, 
    cut_range = axis_chem_shift_range
)

for fold_name, fold_data in closeup_region.items():
    _fig, _ax = Graphs.spectral.spectra(
        x_axis=fold_data.get(CONSTANTS.X_CHEM_SHIFT),
        y_spectra=fold_data.get(CONSTANTS.Y_SPECTRUM),
        label_x="Chemical shift [ppm]",
        title="Close up 3.16 peak region"
    )
    _ax.invert_xaxis()
```
